```python
import random
import pandas
import matplotlib.pyplot as plt
import seaborn
```

```python
        #O seaborn tem vários datasets para utilizar em projetos e para estudo
dados = seaborn.load_dataset("penguins")#Este é um dataset de dados de pinguins
```

```python
print(dados.head())
```

```python
        #data - é a tabela que vai usar de base pro gráfico
seaborn.jointplot(data=dados, x = "bill_length_mm", y = 'bill_depth_mm',  kind='reg')#kind - é o tipo do gráfico, reg é linha de regressão
                #x e y são os eixos do gráfico                                       tem vários tipos de kind

#bill length é o comprimento do bico e bill depth é a altura dele

#Jointplot vai criar gráfico de dispersao junto com o histograma
```

```python
#lmplot é Linear Model Plot, ou gráfico de modelo linear                                          
seaborn.lmplot(data = dados, x = 'bill_length_mm', y = 'bill_depth_mm',  palette={'Male':'blue', #hue divide as cores dos dados com base na coluna 
                                                                                  'Female':'red'}, hue = 'sex', col = 'species', height=3,#'sex'
                                                                                  scatter_kws={'s':4})
#                     o col são os gráficos, que são criados para cada variavel na coluna, no caso aqui é um gráfico para macho e outro pra femea
```

```python
seaborn.lmplot(data = dados, x = 'bill_length_mm', y = 'bill_depth_mm', hue = 'species', col = 'island', height=3, scatter_kws={'s':4})
```

```python
seaborn.kdeplot(x=dados.body_mass_g, hue=dados.species)
```

```python
seaborn.displot(dados.flipper_length_mm)
#Cria marcações para cada valor individual
seaborn.rugplot(dados.flipper_length_mm)
```

```python
seaborn.violinplot(dados.flipper_length_mm)
```

```python
#Criando os dados para a proxima parte
x1 = random.sample(range(0, 100), 50)
y1 = []

for i in x1:
    y1.append(i ** 5)

df = pandas.DataFrame()
df['x1'] = x1
df['y1'] = y1
```

```python
print(df.head(),"\n\n", df.shape)
```

```python
seaborn.lmplot(data = df, x = "x1", y = "y1", fit_reg= False)#fit_reg True é que vai fazer mostrar a linha azul na diagonal do gráfico
```
